## Dataset <a id="0"></a>

The dataset we will be working with is the **"Online Retail"** dataset. It contains transactional data of an online retail store from 2010 to 2011.<br> The dataset is available as a .xlsx file named `Online Retail.xlsx`. This data file can also be downloaded [here](https://archive.ics.uci.edu/ml/machine-learning-databases/00352/Online%20Retail.xlsx).

The dataset contains the following columns:

**InvoiceNo:** Invoice number of the transaction, if the code starts with letter 'c', it indicates a cancellation.<br>
**StockCode:** Unique code of the product.<br>
**Description:** Description of the product.<br>
**Quantity:** Quantity of the product in the transaction.<br>
**InvoiceDate:** Date and time of the transaction.<br>
**UnitPrice:** Unit price of the product.<br>
**CustomerID:** Unique identifier of the customer.<br>
**Country:** Country where the transaction occurred.

## Data Cleaning <a id="3"></a>

**Missing Values**

```python
df.isnull().sum()
```

There are **1454** values in column `'Description'` & **135080** values in `'CustomerID'` column missing, we need to deal with them

**Duplicate Values**

**Dropping Duplicates**

```python
df.drop_duplicates(keep='first', inplace=True, ignore_index=True)
df
```

### Removing Non Sale Entries from the dataframe

```python
# # to view non sale stock codes, uncomment
# df['StockCode'] = df['StockCode'].astype(str)
# pd.DataFrame(df['StockCode'].value_counts().sort_index().tail(40))
```

```python
non_sales_stock_codes = ['AMAZONFEE', 'B', 'BANK CHARGES', 'C2', 'CRUK', 'D', 'DOT', 'POST', 'S', 'M', 'm']

# Filter out non-sales entries
sales_df = df[~df['StockCode'].isin(non_sales_stock_codes)]
sales_df
```

There were `2906 non sale entries` and they have been removed and a new dataframe "sales_df", containing only sales entries is created

**Dropping entries with 'UnitPrice' == 0**

```python
sales_df[sales_df['UnitPrice'] == 0].count()
```

There are `2496 entries` with unitprice equals 0, they can be removed 

```python
sales_df = sales_df[~(sales_df['UnitPrice'] == 0)].reset_index(drop=True)
```

**Missing Values Dealt**

```python
sales_df.isnull().sum()
```

**Explanation:**

By Filtering out non sale entries, duplicates and dropping unit price equals 0 entries.<br> All the missing values in the column `'Description'` were removed.<br> Now, we still have 131580 missing values in the `'CustomerID'` column, these cannnot be filled using any criteria. It needs to be considered as per the requirement later

**Changing the order of the columns**

```python
sales_df = sales_df[['CustomerID', 'InvoiceNo', 'StockCode', 'UnitPrice', 'Quantity', 'InvoiceDate', 'Description', 'Country']]
```

**Changing the 'CustomerID' Data type**

```python
sales_df.loc[:, 'CustomerID'] = sales_df['CustomerID'].astype('Int64')
```

# Exploring Specific Questions <a id="5"></a>

```python
# Creating a new dataframe to calculate total sales 
revenue_df = sales_df.copy()
TotalAmount = revenue_df['Quantity'] * revenue_df['UnitPrice']
revenue_df.insert(loc=5,column='TotalAmount',value=TotalAmount)
revenue_df
```

## 1. Total Sales Revenue Across Countries

```python
# grouping data by 'Country' and summing the 'TotalAmount'
country_totals = revenue_df.groupby('Country')['TotalAmount'].sum().sort_values(ascending=False)

# plotting a bar graph
country_totals.plot(kind='bar', figsize=(15,7))

plt.title('\n Total Sales Revenue Across Countries \n',fontdict={'fontsize' : 18})
plt.xlabel('\n Country \n', fontdict={'fontsize' : 14})
plt.ylabel('\n Total sales \n',fontdict={'fontsize' : 14})

# setting x and y ticks
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# setting y axis tick labels to non scientific notation
plt.ticklabel_format(style='plain', axis='y')

# Annotating first 5 bars with their values
for i, v in enumerate(country_totals[:5]):
    plt.annotate(str(round(v, 2)), (i, v), ha='center', va='bottom', rotation=60, fontsize=10)

# setting x and limits
plt.xlim(-1,)
plt.ylim(0, 10000000)

plt.grid(False)
plt.show()
plt.close()
```

**Inference:**

1. We can see that **United Kingdom** brings in a major chunk of sales, 8282076 **(8.28 Millions approx)**.
2. It is followed by **Netherlands & EIRE** but the figures are meagre.



## 2a. Top 10 Revenue Contributing Counties

```python
# grouping data by 'Country' and summing the 'TotalAmount', filtering the top 10
country_top10 = revenue_df.groupby('Country')['TotalAmount'].sum().sort_values(ascending=False)[0:10]

# plotting a bar graph
country_top10.plot(kind='bar', figsize=(10,7))

plt.title('\n Top 10 Revenue Contributing Countries \n',fontdict={'fontsize' : 18})
plt.xlabel('\n Country \n', fontdict={'fontsize' : 14})
plt.ylabel('\n Total sales \n',fontdict={'fontsize' : 14})

# setting x and y ticks
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# setting y axis tick labels to non scientific notation
plt.ticklabel_format(style='plain', axis='y')

# Annotating bars with their values
for i, v in enumerate(country_top10):
    plt.annotate(str(round(v, 2)), (i, v), ha='center', va='bottom', rotation=60, fontsize=10)

# setting x and limits
plt.ylim(0, 10000000)

plt.grid(False)
plt.show()
plt.close()
```

**Inference:**

1. We can easily say that **United Kingdom** is the biggest, perhaps only significant contributor to sales revenue.
2. The others in the list do not contribute much.
3. Apart from **United Kingdom**, the three other markets that can be considered are **Netherlands, EIRE, & Germany**.<br>However, they are in a totally different frame.

## 3. Total Sales Across Months

```python
# creating more columns 'WeekDay','Hour, 'Day', 'Month' and 'Year' from 'InvoiceDate' column
revenue_df['WeekDay'] = revenue_df['InvoiceDate'].dt.day_name()
revenue_df['Hour'] = revenue_df['InvoiceDate'].dt.hour
revenue_df['Day'] = revenue_df['InvoiceDate'].dt.day
revenue_df['Month'] = revenue_df['InvoiceDate'].dt.month
revenue_df['Year'] = revenue_df['InvoiceDate'].dt.year
```

```python
# grouping data by 'Year', 'Month', summing the 'TotalAmount' to get monthly sales
monthly_sales = revenue_df.groupby(['Year', 'Month'])['TotalAmount'].sum()

# plotting a bar graph
monthly_sales.plot(kind='bar', figsize=(10,7))

plt.title('\n Total Sales Across Months \n',fontdict={'fontsize' : 18})
plt.xlabel('\n Month, Year \n', fontdict={'fontsize' : 14})
plt.ylabel('\n Total sales \n',fontdict={'fontsize' : 14})

# setting x and y ticks
xtick_labels = ['Dec_10', 'Jan_11', 'Feb_11', 'Mar_11', 'Apr_11', 'May_11', 'Jun_11', 'Jul_11', 'Aug_11', 'Sep_11', 'Oct_11', 'Nov_11', 'Dec_11']
plt.xticks(ticks=range(13), labels=xtick_labels,rotation=0, fontsize=10)
plt.yticks(fontsize=10)

# setting y axis tick labels to non scientific notation
plt.ticklabel_format(style='plain', axis='y')

# Annotating bars with their values
for i, v in enumerate(monthly_sales):
    plt.annotate(str(round(v, 2)), (i, v), ha='center', va='bottom', rotation=60, fontsize=10)
    
# setting x and limits
plt.ylim(0, 1800000)

plt.show()
plt.close()
```

**Inference:**

  1. Knowing the fact that we do not have complete data for the month of Dec_2011,<br> we can say that **minimum sales were achieved in Apr_2011** and **maximum were achieved in Nov_2011**.
  2. After the dip in Apr_11, there is a spike in sales observed from May_11.<br> Although, there was a slow down in between, there is a gradual and considerable **increasing trend in sales**.

## 5a. Top 10 Products sold

```python
pd.DataFrame(sales_df['Quantity'].describe(percentiles=[0.1, 0.25, 0.5, 0.75, 0.99, 0.9999])).T
```

We can see that **99.99%** of the product quantity sold lies below '1411' no's.<br>Hence, lets for the case here, lets filter out negative quantities and quantity values above 1500.

```python
# filtering negtive quantities and quantity above 1500
sales_filtered_df = sales_df[(sales_df['Quantity'] > 0) & (sales_df['Quantity'] < 1500)]

# creating a pivot table by indexing 'StockCode' and 'Description' to find the top 10 products sold by quantity
top10_products = pd.pivot_table(sales_filtered_df,
    index=['StockCode','Description'],
    values='Quantity',
    aggfunc='sum').sort_values(by='Quantity', ascending=False)
print('\n Top 10 Products Sold: \n')
top10_products.head(10)
```

```python
# Reset the index of the pivot table created for plotting
top10_products.reset_index(inplace=True)

# plotting a horizontal bar graph by reversing the order of rows
top10_products.head(10).iloc[::-1].plot(kind='barh', y='Quantity', x='Description', legend=False, figsize=(10,7))

plt.title('\n Top 10 Products Sold \n', fontdict={'fontsize' : 18})
plt.xlabel('\n Quantity \n',fontdict={'fontsize' : 14} )
plt.ylabel('')

# setting x and y ticks
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# showing value of each bar
for index, value in enumerate(top10_products.head(10)['Quantity'].iloc[::-1]):
    plt.annotate(str(value), xy=(value, index), ha='left', va='center', fontsize=10)


# setting x limits
plt.xlim(0, 55000)

plt.grid(False)
plt.show()
plt.close()
```

**Inference:**

  1. It can be observed that the **Jumbo Bag Red Retrospot** is the most sold with numbers 48371.
  2. The list also conveys the 9 other products.<br> Steps can be taken to ensure constant **supply and storage** of these products.
  3. It can be evaluated if these products can also be promoted in other markets appropriately.

## 10. No of Orders for Different Days of Week

```python
# Using groupby to count no. of Invoices daywise
orders_weekdays = revenue_df.groupby('InvoiceNo')['WeekDay'].unique().value_counts().sort_index()

orders_weekdays.plot(kind='bar',figsize=(15,6))

plt.title('\n Number of orders for Different Days of Week \n', fontdict={'fontsize' : 18})
plt.xlabel('\n Day of Week \n', fontdict={'fontsize' : 14})
plt.ylabel('\n Number of Orders \n', fontdict={'fontsize' : 14})

# x and y ticks
xtick_labels = ['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sun']
xticks = range(len(orders_weekdays.index))
plt.xticks(ticks=xticks, labels=xtick_labels , rotation=0, fontsize=10)
plt.yticks(fontsize=10)

# setting y limits
plt.ylim(0,6000)

plt.show()
plt.close()
```

**Inference:**

  1. **Thursday** is the day which brings in majority of sales.
  2. There is no data for **Saturday** suggesting store is closed on Saturdays.
  3. Also, we can see that sales are more across weekends.
  4. Since, the least sales come from **Wednesday**, the store can have a weekly off on Wednesday instead of Saturday.<br> This will insure sales are **continuous through the weekend**.
